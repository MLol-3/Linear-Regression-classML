```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('HeightWeight20.csv')

# Extract the features and target
X = data[['Weight']].values
y = data['Height'].values

# Initialize parameters
m = 0  # slope
c = 0  # intercept
learning_rate = 0.01
num_iterations = 1000

# List to store the SSR for each iteration
ssr_list = []

# Number of samples
n = len(y)

# Gradient Descent
for i in range(num_iterations):
    # Predict the output with the current parameters
    y_pred = m * X + c

    # Calculate the residuals
    residuals = y_pred - y.reshape(-1, 1)

    # Calculate the sum of squared residuals (SSR)
    ssr = np.sum(residuals**2)
    ssr_list.append(ssr)

    # Calculate gradients
    dm = (-2/n) * np.sum(X * residuals)
    dc = (-2/n) * np.sum(residuals)

    # Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc

    # Print progress every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}: SSR = {ssr}, m = {m}, c = {c}")

# Final model parameters
print(f"Final parameters: m = {m}, c = {c}")

# Plotting the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, m * X + c, color='red', label='Regression line')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Linear Regression without Library')
plt.legend()
plt.show()

# Plotting SSR over iterations
plt.plot(range(num_iterations), ssr_list, color='green')
plt.xlabel('Iteration')
plt.ylabel('SSR')
plt.title('SSR Over Iterations')
plt.show()
```
